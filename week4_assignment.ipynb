This cell imports the required Python libraries for data analysis and visualization.
pandas is used to read and manipulate the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

The dataset student_scores.csv is loaded into a DataFrame called df, and displaying it helps verify that the data has been read correctly.

```python
df = pd.read_csv("/content/student_scores (1).csv")
df
```

```python
df.describe()
```

```python
plt.scatter(df["Hours"], df["Scores"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study Time vs Exam Scores")
plt.show()
```

This histogram shows how study hours are distributed among students.
It helps identify:
  1.Common study time ranges
  2.Skewness or clustering in the data
Understanding this distribution is important for interpreting correlation and hypothesis test results.

```python
plt.hist(df["Hours"])
plt.xlabel("Study Hours")
plt.title("Distribution of Study Hours")
plt.show()
```

This histogram displays the frequency distribution of exam scores.
It provides insight into:

Performance spread among students

Whether scores are concentrated at lower or higher ranges
This visualization supports understanding score variability.

```python
plt.hist(df["Scores"])
plt.xlabel("Exam Scores")
plt.title("Distribution of Exam Scores")
plt.show()
```

```python
'''This cell calculates the Pearson correlation coefficient and p-value.
Correlation coefficient (r) measures the strength and direction of the linear relationship.
p-value indicates whether the observed correlation is statistically significant.
A high positive correlation close to +1 indicates a strong positive relationship between study time and exam scores.
'''
correlation, p_value = stats.pearsonr(df["Hours"], df["Scores"])
correlation, p_value
```

```python
alpha = 0.05

if p_value < alpha:
    "Reject Null Hypothesis: Study time significantly affects exam scores"
else:
    "Fail to Reject Null Hypothesis: No significant effect found"
```

```python
slope, intercept, r, p, std_err = stats.linregress(df["Hours"], df["Scores"])
slope, intercept, r, p
```

This cell performs simple linear regression to model the relationship between study hours and exam scores.

Slope shows how much the score increases per additional hour of study.

Intercept is the expected score when study hours are zero.

r-value and p-value confirm the strength and significance of the model.

Regression provides predictive insight beyond correlation.

```python
plt.scatter(df["Hours"], df["Scores"])
plt.plot(df["Hours"], intercept + slope * df["Hours"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Linear Regression: Study Time vs Scores")
plt.show()
```

Based on correlation analysis, visualization, and hypothesis testing, the study concludes that study time has a statistically significant and positive impact on exam scores. Students who invest more hours in studying tend to achieve higher exam results.
